```python
def measure(img, ROI):
  
  x1, x2, y1, y2 = ROI
  ROI = img[x1:x2, y1:y2]
  print('the mean value is ' + str(ROI.mean()))
  print('the max value is ' + str(ROI.max()))
  print('the min value is ' + str(ROI.min()))
  print('the size is ' + str(ROI.shape))
  print('the stdev value is ' + str(ROI.std()))
```

```python
def display(img_stack, ROI, slice_num):
  #E2_image_stack,ROI, slice_num
  x1, x2, y1, y2 = ROI

  value = np.shape(slice_num)

  fig = plt.figure(figsize=(15, 10))
  for i in range (0, value[0]):
    temp_img = img_stack[:,:,slice_num[i]]
    ROI = temp_img[x1:x2, y1:y2]
    #signal_values.append(ROI.mean())
    plt.subplot(1, 3, i+1)
    plt.imshow(temp_img, cmap='gray')
    plt.gca().add_patch(Rectangle((x1,y1),(x2-x1),(y2-y1),linewidth=1,edgecolor='r',facecolor='none'))
    plt.axis('off')
```

```python
def mean_value(img_stack, ROI, slice_num_1):
  x1, x2, y1, y2 = ROI
  #find the mean value of the ROI
  a = img_stack[:,:,slice_num_1]
  ROI = a[x1:x2, y1:y2]  
  return ROI.mean()
```

```python
titles = ['Approximation', ' Horizontal detail',
          'Vertical detail', 'Diagonal detail']
coeffs2 = pywt.dwt2(ds.pixel_array, 'bior1.3')
LL, (LH, HL, HH) = coeffs2
fig = plt.figure(figsize=(12, 3))
for i, a in enumerate([LL, LH, HL, HH]):
    ax = fig.add_subplot(1, 4, i + 1)
    ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
    ax.set_title(titles[i], fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

fig.tight_layout()
plt.show()
```
